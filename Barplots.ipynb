```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = np.arange(10,20)

x = np.arange(0,10) + np.random.normal(size=x.size)
y = np.arange(10,20) + np.random.normal(size=y.size)

fig, ax = plt.subplots()
plt.scatter(x, y, color="red")

a, b = np.polyfit(x,y,1)
plt.plot(x, a*x+b, color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.text(2, 18, 'y = '+'{:.2f}'.format(b)+' +{:.2f}'.format(a)+'x')
plt.show()
```

```python
fig, ax = plt.subplots() 
langs = ['C', 'C++', 'Java', 'Python', 'PHP'] 
students = [23,17,35,29,12] 
yerr1 = [0.5, 4, 0.5, 1, 10]

plt.bar(langs,students, yerr=yerr1, capsize=2, color='green', edgecolor = 'black', width=0.2) 
plt.show()
```

```python
barWidth = 0.3 
 
bars1 = [10, 9, 2] 
bars2 = [10.8, 9.5, 4.5] 
  
yer1 = [0.5, 0.4, 0.5] 
yer2 = [1, 0.7, 1] 
  
r1 = np.arange(len(bars1)) 
r2 = [x + barWidth for x in r1] 
  
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', yerr=yer1, capsize=2, label='one') 
plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', yerr=yer2, capsize=2, label='two')  

plt.xticks([r + barWidth for r in range(len(bars1))], ['A', 'B', 'C']) 

plt.ylabel('height') 
plt.legend() 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

company = (['GOOGL', 'AMZN', 'MSFT', 'FB'])
revenue = ([90, 136, 89, 27])

fig, ax=plt.subplots()
x_pos = np.arange(len(company))
plt.xticks(x_pos, company)
plt.bar(x_pos, revenue, width=0.4, color='pink', label='revenue')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

data = {'facebook':100, 'amazon':250, 'youtube':89, 'twitter':78}
x = list(data.keys())
y = list(data.values())

#fig, ax = plt.subplots()
#style.use('ggplot')
plt.figure(figsize=(5,5))
x_pos = np.arange(len(x))
plt.xticks(x_pos, x)
color = ['red', 'pink', 'blue', 'green']

plt.bar(x_pos, y, color=color, width=0.4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

fb = [120, 35, 60, 98]
ig = [38, 54, 93, 89]
yt = [45, 78, 12, 90]
tw = [90, 45, 11, 76]

#fig, ax = plt.subplots()
#plt.figure(figsize=(10,7))
x_pos = np.arange(len(fb))

plt.bar(x_pos, fb, color='red', width=0.2, label='Facebook')
plt.bar(x_pos, ig, bottom=fb, color='green', width=0.2, label='Instagram')
plt.bar(x_pos+0.2, yt, color='orange', width=0.2, label='Youtube')
plt.bar(x_pos+0.4, tw, color='blue', width=0.2, label='Twitter')

plt.xticks(x_pos+0.2, ('Monday', 'Tuesday', 'Wednesday', 'Thursday'))


plt.legend()
plt.show()
```

```python
import numpy as np
from matplotlib import style
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)


x = np.linspace(0,15,30)
y = np.sin(x) + 0.1*np.random.randn(len(x))

plt.figure(figsize=(8,3))

plt.subplot(1,2,1)
plt.plot(x,y)
dydx = np.gradient(y, x)
plt.plot(x, dydx)

plt.subplot(1,2,2)
plt.plot(x,y, 'o', color='red')
plt.show()
```
